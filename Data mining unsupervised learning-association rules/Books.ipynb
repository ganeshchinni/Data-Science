```python
final_books=books.drop_duplicates()
final_books
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#calculating frequent items
frequent_items=apriori(books,min_support=0.0075,max_len=4,use_colnames=True)
frequent_items
```

```python
#most frequent items based on support
frequent_items.sort_values('support',ascending=False,inplace=True)
frequent_items
```

```python
#Association rules: lift
rules=association_rules(frequent_items,metric='lift',min_threshold=1)
rules
```

```python
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]).head(10)
rules
```

```python
#Association rules: cofidence
rules=association_rules(frequent_items,metric='confidence',min_threshold=1)
rules.head(10)
```
