```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
hault=pd.read_csv('/content/drive/MyDrive/data science/flight_hault.csv')
routes=pd.read_csv('/content/drive/MyDrive/data science/connecting_routes.csv')
```

```python

hault.columns=["ID","Name","City","Country","IATA_FAA","ICAO","Latitude","Longitude","Altitude","Time","DST","Tz database time"]
```

```python
hault.describe()
```

```python

routes.columns=["flights", " ID", "main_Airport","main_Airport_ID", "Destination ","Destination  ID","haults","hault","machinary"]
```

```python
hault_routes=pd.concat([hault, routes], axis=1)
g=nx.Graph()
g = nx.from_pandas_edgelist(hault_routes, source = "main_Airport", target = "Destination ")
print(nx.info(g))
```

```python
#degree of centrality
b=nx.degree_centrality(g)
print(b)
```

```python

pos = nx.spring_layout(g, k = 0.15)
nx.draw_networkx(g, pos, node_size = 25, node_color = 'blue')
```

```python
#closeness centrality
closeness=nx.closeness_centrality(g)
print(closeness)
```

```python
#beteweness centrality

b_cent = nx.betweenness_centrality(g)
print(b_cent)
```
